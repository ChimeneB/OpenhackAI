```python
import pprint
import requests
from io import BytesIO
from PIL import Image
from postmen import Postmen, PostmenException

# https://pypi.org/project/postmen/0.1/
# https://github.com/postmen/postmen-sdk-python
# https://docs.postmen.com/
```

```python
url = 'https://images-na.ssl-images-amazon.com/images/I/81pr3S9d8lL._UL1500_.jpg'

response = requests.get(url)
img = Image.open(BytesIO(response.content))
img
```

```python
pp = pprint.PrettyPrinter(indent=4)

# how get API key and region https://docs.postmen.com/

api_key = 'YOUR_API_KEY'
region = 'sandbox'

# create Postmen API handler object

api = Postmen(api_key, region)

try:
    # as an example we request all the labels

    result = api.get('labels')
    print("RESULT:")
    pp.pprint(result)
except PostmenException as e:
    # if error occurs we can access all
    # the details in following way

    print("ERROR")
    print(e.code())  # error code
    print(e.message())  # error message
    pp.pprint(e.details())  # details
```
